**4. Produce a histogram of the penguins' _body mass_ measurements with 12 bins.**  You could do this by using (a) the Pandas API, or (b) the Seaborn API. **5. Produce a histogram of _body mass_ that also conditions on _sex_.** In other words, play around with visualizations that show the distribution of body mass separately for male and female penguins.

There are different ways to do this: one is by using different colors (hues) in the histogram to distinguish male and female penguins. Seaborn makes this really, really easy! Focus on the Seaborn tutorial pages to figure out how to do this, and try out the different options Seaborn provides to combine the histogram bars (e.g., "stacking" vs. "layering" vs. "dodging"). Which option, in your opinion, provides the most useful visualization in this case?

```python
# Your code here
sns.histplot(data=df, x="body_mass_g", hue="sex")
```

```python
sns.histplot(data=df, x="body_mass_g", hue="sex", multiple='stack')
```

```python
sns.histplot(data=df, x="body_mass_g", hue="sex", multiple='layer')
```

```python
sns.histplot(data=df, x="body_mass_g", hue="sex", multiple='dodge')
```

```python
# Your code here - multiple linear regressions (conditioning on the species)
sns.lmplot(data=df, x="bill_length_mm", y="body_mass_g", hue='species');
```
